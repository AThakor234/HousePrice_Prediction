```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = "C:/Users/aadiv/Housing_Project/Dataset/Housing_Project_Dataset.csv"
house_df = pd.read_csv(file_path)
house_df.head(20)
```

```python
#drop repititive columns
house_df = house_df.drop(["Heating features",'Cooling features','Parking features'], axis=1)
```

```python
house_df.info()
```

```python
# Determine the number of unique values in each column.
house_df.nunique()
```

```python
# define columns with null values
features_with_na = [features for features in house_df.columns if house_df[features].isnull().sum()>1]
features_with_na
```

```python
#Define columns with numerical values that have null values

numerical_with_nan=[feature for feature in house_df.columns if house_df[feature].isnull().sum()>1 and house_df[feature].dtypes!='O']

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.round(house_df[feature].isnull().mean(),4)))
```

```python
# Replace the null values with the mean 

for feature in numerical_with_nan:
  mean_value=house_df[feature].mean()
  house_df[feature].fillna(mean_value,inplace=True)
    
house_df[numerical_with_nan].isnull().sum()
```

```python
#Define columns with categorical variables that have null values

categorical_nan=[feature for feature in house_df.columns if house_df[feature].isnull().sum()>1 and house_df[feature].dtypes=='O']

for feature in categorical_nan:
    print("{}: {}% missing values".format(feature,np.round(house_df[feature].isnull().mean(),4)))
```

```python
# Replace the null values with the string "Missing"

for feature in categorical_nan:
  house_df[feature].fillna('Missing',inplace=True)

house_df[numerical_with_nan].isnull().sum()
```

```python
#check for null values
house_df.isna().sum()
```

```python
#display dataframe
house_df.head(20)
```

```python
#check dataframe datatypes
house_df.dtypes
```

```python
#convert corresponding columns to integer datatype
house_df['Year built'] = house_df['Year built'].astype(int)
house_df['Elementary School Score'] = house_df['Elementary School Score'].astype(int)
house_df['Middle School Score'] = house_df['Middle School Score'].astype(int)
house_df['High School Score'] = house_df['High School Score'].astype(int)
house_df['Bathrooms'] = house_df['Bathrooms'].astype(int)
house_df['Full bathrooms'] = house_df['Full bathrooms'].astype(int)
house_df['Full bathrooms'] = house_df['Full bathrooms'].astype(int)
house_df['Garage spaces'] = house_df['Garage spaces'].astype(int)
house_df['Total spaces'] = house_df['Garage spaces'].astype(int)
```

```python
#convert Listed on column to datetime datatype
house_df['Listed On'] = pd.to_datetime(house_df['Listed On'])
```

```python
#replace the missing values in the Last sold on column with a generic datae and convert to datetime datatype
house_df['Last Sold On'] = house_df['Last Sold On'].str.replace('Missing','2019-08-30')
house_df['Last Sold On'] = pd.to_datetime(house_df['Last Sold On'])
```

```python
#check Bedroom column value count
house_df["Bedrooms"].value_counts()
```

```python
#perform a split on the bedroom column and pick the first element only
for i in range(len(house_df)):
    house_df.loc[i,"Bedrooms"] = house_df.loc[i,"Bedrooms"].split()[0]
```

```python
#check Bedroom column again for unique value count
house_df['Bedrooms'].value_counts()
```

```python
# drop rows with inappropriate values in Bedroom column
values_to_drop = ['Master','Missing','Ground','Walk-in','More','Loft','Inverted', 'Reverse']

for value in values_to_drop:
    house_df = house_df[~house_df['Bedrooms'].str.contains(value)]
```

```python
#convert Bedroom column to integer datatype
house_df['Bedrooms'] = house_df['Bedrooms'].astype(int)
```

```python
#check dataframe datatypes
house_df.dtypes
```

```python
house_df.rename(columns={'Id':'Housing_Id'}, inplace=True)
house_df.rename(columns={'Type':'Housing_type'}, inplace=True)
house_df.rename(columns={'State':'H_State'}, inplace=True)
house_df.rename(columns={'Year built':'Year_built'}, inplace=True)
house_df.rename(columns={'Elementary School Score':'elementary_school_score'}, inplace=True)
house_df.rename(columns={'Elementary School Distance':'elementary_school_distance'}, inplace=True)
house_df.rename(columns={'Middle School Score':'middle_school_score'}, inplace=True)
house_df.rename(columns={'Middle School Distance':'middle_school_distance'}, inplace=True)
house_df.rename(columns={'High School Score':'high_school_score'}, inplace=True)
house_df.rename(columns={'High School Distance':'high_school_disctance'}, inplace=True)
house_df.rename(columns={'Parking':'parking'}, inplace=True)
house_df.rename(columns={'Bedrooms':'bedrooms'}, inplace=True)
house_df.rename(columns={'Bathrooms':'bathrooms'}, inplace=True)
house_df.rename(columns={'Full bathrooms':'full_bathrooms'}, inplace=True)
house_df.rename(columns={'Total interior livable area':'total_interior_livable_area'}, inplace=True)
house_df.rename(columns={'Heating':'heating_features'}, inplace=True)
house_df.rename(columns={'Cooling':'cooling_features'}, inplace=True)
house_df.rename(columns={'Tax assessed value':'tax_assesses_value'}, inplace=True)
house_df.rename(columns={'Annual tax amount':'annual_tax_amount'}, inplace=True)
house_df.rename(columns={'Listed Price':'listed_price'}, inplace=True)
house_df.rename(columns={'Last Sold Price':'last_sold_price'}, inplace=True)
house_df.rename(columns={'Sold Price':'sold_price'}, inplace=True)
house_df.rename(columns={'Listed On':'Listed_on'}, inplace=True)
```
