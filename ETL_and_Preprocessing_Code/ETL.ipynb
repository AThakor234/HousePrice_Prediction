```python
#drop repititive columns
house_df = house_df.drop(["Heating features",'Cooling features','Parking features'], axis=1)
```

```python
# define columns with null values
features_with_na = [features for features in house_df.columns if house_df[features].isnull().sum()>1]
features_with_na
```

```python
#Define columns with numerical values that have null values
numerical_with_nan=[feature for feature in house_df.columns if house_df[feature].isnull().sum()>1 and house_df[feature].dtypes!='O']
for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(house_df[feature].isnull().mean(),4)))
```

```python
# Replace the null values with the mean 

for feature in numerical_with_nan:
  mean_value=house_df[feature].mean()
  house_df[feature].fillna(mean_value,inplace=True)
house_df[numerical_with_nan].isnull().sum()
```

```python
#Define columns with categorical variables that have null values
categorical_nan=[feature for feature in house_df.columns if house_df[feature].isnull().sum()>1 and house_df[feature].dtypes=='O']

for feature in categorical_nan:
    print("{}: {}% missing values".format(feature,np.round(house_df[feature].isnull().mean(),4)))
```

```python
# Replace the null values with the string "Missing"

for feature in categorical_nan:
  house_df[feature].fillna('Missing',inplace=True)
house_df[numerical_with_nan].isnull().sum()
```
